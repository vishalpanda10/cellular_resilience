```python
def get_gpt_embedding(text, model="text-embedding-ada-002"):
    text = text.replace("\n", " ")
    return np.array(openai.Embedding.create(input = [text], model=model)['data'][0]['embedding'])
```

```python
file_path = '../../../data/BreastCancer10xGenomics_Rep1/exported_data/gene_data_from_ncbi_genept.json'

gene_data = json.load(open(file_path, 'r'))
```

```python
gpt_embeddings = {}
```

```python
GPT_DIM = 1536 
for key, text in sorted(gene_data.items()):
    if key not in gpt_embeddings:
        print('key',key)
        if gene_data[key] == '': 
            gpt_embeddings[key] = np.zeros(GPT_DIM) 
        else:
            gpt_embeddings[key] = get_gpt_embedding(text)
```

```python
with open('../../../data/BreastCancer10xGenomics_Rep1/exported_data/gene_embeddings.pickle', "wb") as file:
    pickle.dump(gpt_embeddings, file)
```
